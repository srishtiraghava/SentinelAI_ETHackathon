```python
from mftool import Mftool
import pandas as pd

mf = Mftool()
scheme_code = '118989' # HDFC Mid-Cap Direct Growth

# 1. Try to get the live quote
quote = mf.get_scheme_quote(scheme_code)
nav = quote.get('nav')

# 2. Fallback Logic: If NAV is None (Weekends/Holidays), get the last historical NAV
if nav is None:
    # Fetch historical data and convert to a DataFrame
    history = mf.get_scheme_historical_nav(scheme_code, as_Dataframe=True)
    if not history.empty:
        # Get the first row (the most recent update)
        nav = history.iloc['nav']

print(f"Fund: {quote.get('scheme_name')}")
print(f"Status: Using {'Live' if quote.get('nav') else 'Last Available'} Data")
print(f"NAV: ₹{nav}")
```
